# Arithmetic Series

## Modified from Arithmetic Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/ArithmeticSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Arithmetic_progression)


An arithmetic sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous one by adding (or subtracting) a constant $d$.

\begin{align*}
a_1 & \\
a_2 &\ \ =\ \ a_1 + d  \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d   \\
a_4 &\ \ =\ \ a_3 + d \ \ =\ \ a_2 + 2d \ \ =\ \ a_1 + 3d  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ =\  \ a_1+(k-1)d
\end{align*}

Therefore, for $k>=1$,

$$ a_k = a_1 + (k-1)d$$

With $a_1$ and $d$ given.

<br />

The sequence of the partial sum of the first $n$ terms of a sequence is called a series, in the case of an arithemetic sequence the series is called an arithmetic series.

If  $a_1= 1$ and $d=1$ we have, the simplest arithmnetic sequence,  the natural numbers sequence $\{a_k = k \}$ for  $k=1, 2, ..., $  i.e. the seuqnece
$$1, 2, ...,$$
The  corresponding series is
$$
S_n = 1 + 2+ \ldots + (n-1)+ n = \sum_{k=1}^n k
$$
can be calculated as a function of $n$.

Assume that we allign one $S_{n-1}$ with a second  $S_{n-1}$ that have the terms adden reverse order and you add term to term, we have
\begin{array}{llclclclclc}
  S_{n-1} &=& 1 &+& 2 &+& \ldots &+& (n-2) &+& (n-1) \\
  S_{n-1} &=& (n-1)  &+& (n-2)  &+& \ldots   &+& 2   &+& 1 \\
   \hline \\
   2S_{n-1} &=& n  &+& n  &+& \ldots   &+& n   &+& n \\
\end{array}
so
$$  2S_{n-1} = (n-1)n $$
then
$$  S_{n-1} = \frac{1}{2} (n-1)n$$
therefore
$$  S_{n} = \frac{1}{2} n(n+1)$$


\begin{align*}
S_n &= \sum_{k=1}^n a_k	\\
    &= \sum_{k=1}^n[a_1+(k-1)d]		\\
    &= na_1 + d \sum_{k=1}^n(k-1)	\\
    &= na_1 + d \sum_{k=0}^{n-1} k	\\
    &= na_1 + d \sum_{k=1}^{n-1} k	\\
    &= na_1 +  \frac{1}{2} d(n-1)n\\
\end{align*}
so
$$ S_n =  \frac{1}{2} n[2 a_1 +d(n-1)].$$

Note, however, that
$$a_1+a_n=a_1+[a_1+d(n-1)]=2a_1+d(n-1),$$
so
$$S_n = \frac{1}{2}n(a_1+a_n),$$


or $n$ times the arithmetic mean of the first and last terms! This is the trick Gauss used as a schoolboy to solve the problem of summing the integers from $1$ to $100$ given as busy-work by his teacher. While his classmates toiled away doing the addition longhand, Gauss wrote a single number, the correct answer

$$\frac{1}{2}(100)(1+100)=50Â·101=5050$$
on his slate (Burton 1989, pp. 80-81; Hoffman 1998, p. 207). When the answers were examined, Gauss's proved to be the only correct one.

\begin{align*}
a_0 & \\
a_1 &\ \ =\ \ a_0 + d  \\
a_2 &\ \ =\ \ a_1 + d \ \ =\ \ a_0 + 2d   \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d \ \ =\ \ a_0 + 3d  \\

a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ = \  \ a_1+(k-1)d \ \ =\ \ a_0+kd
\end{align*}

for given   $a_0$ and $d$.

In this case the sum of the terms between $0$ and $n$ of  the arithemnetic sequence  is then given by
\begin{align*}
S_n &= \sum_{k=0}^{n} a_k	\\
    &= \sum_{k=0}^{n}[a_0+kd]		\\
    &= (n+1) a_0 + d \sum_{k=0}^{n} k	\\
    &= (n+1) a_0 + d \sum_{k=1}^{} k	\\
    &=(n+1) a_0+  \frac{1}{2} dn(n+1) \\
\end{align*}
so
$$ S_n =  \frac{1}{2} (n+1)[2 a_0 + dn].$$

## My own example of arithmetic series

The following code computes the $n=150000$ term of the arithmetic series. The general formula for the $k$-th term $a_k$ of the arithmetic series is:

$$a_k = a + d \cdot (k-1)$$

where:
- $a$ is the first term of the series, and in this case $a = 5$.
- $d$ is the common difference between consecutive terms, which is set as $d = 7$.

Following this formula, the first terms of the arithmetic series are:

$$5, 12, 19, 26, \ldots$$

Each term $a_k$ is obtained by adding the common difference $d$ to the previous term.

Additionally, the code also calculates the accumulated sum $S_n$ of the terms, which is computed using:
$$S_k = \sum_{i=1}^k a_i$$
Where $a_i$ is the $i$-th term of the arithmetic series.

```
import matplotlib.pyplot as plt

# DefiniciÃ³n de variables
n = 150000
a = 5
d = 7
S_n = 0
Serie_S_n = []
Sequence_a_n = []

# CÃ¡lculo de la serie y la suma acumulada
for i in range(1, n+1):
    a_n = a + d * (i-1)
    S_n = S_n + a_n
    Serie_S_n.append(S_n)
    Sequence_a_n.append(a_n)

# GraficaciÃ³n
plt.figure(figsize=(10,6))
plt.title("Serie aritmÃ©tica y suma acumulada")
plt.xlabel("TÃ©rmino n")
plt.ylabel("Valor")
plt.plot(Serie_S_n, color='blue', label='Suma acumulada S_n')
plt.plot(Sequence_a_n, color='green', linestyle='--', label='Serie aritmÃ©tica a_n')
plt.legend()
plt.grid(True)
plt.show()
```

# Geometric Series

## Modified from Geometric Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/GeometricSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Geometric_series)


An geometric series is the sum of a sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous  by multipliying constant $r$ and starts with 1.

\begin{align*}
a_0 &\ \ =\ \ r^0 \ \ =\ \ 1\\
a_1 &\ \ =\ \ a_0 *  r \ \ =\ \  r\\
a_2 &\ \ =\ \ a_1 * r\ \ =\ \ a_0 * r^2  \ \ =\ \ r^2 \\
a_3 &\ \ =\ \ a_2 * r \ \ =\ \ a_0 * r^3 \ \ =\ \  r^3  \\
a_4 &\ \ =\ \ a_3 * r\ \  =\ \ a_0 * r^4 \ \  =\ \ r^4  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}* r \ =\   a_0 * r^k \ \  =\ \  r^k
\end{align*}

The sum of the first $n$ terms  of  the geometric sequence (called an geometric serie)  is then given by
$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  (1 + r + r^2 + \cdots + r^{n-1} ) +r^n \\
 &= S_{n-1}  +r^n
\end{array}
$$

$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  1 + (r + r^2 + \cdots + r^{n-1}  +r^n)  \\
 &=   1 + r (1 +  r  + r^2 + \cdots + r^{n-1} )  \\
  &=   1 + r   S_{n-1}
\end{array}
$$

Then
$$
 S_{n-1}  +r^n =   1 + r   S_{n-1}
$$

Therefore
$$
 S_{n-1} (1 -r)   +r^n =   1 - r^n
$$

$$
 S_{n-1}   =   \frac{1 - r^n }{1 -r}
$$

$$
 S_{n}   =   \frac{1 - r^{n+1} }{1 -r}  = \frac{ r^{n+1} - 1}{r -1}
$$

## Add the a geometric series with $r =2$ up to 10** 6 and verify with the formula.  

```
import matplotlib.pyplot as plt

# Definitions
a = 1
r1 = 2
limit = 10**6

# Function to generate terms and partial sums until r^n exceeds the limit
def generate_series_until_limit(a, r, limit):
    terms = []
    n = 0
    while a * r**n < limit:
        terms.append(a * r**n)
        n += 1
    partial_sums = [sum(terms[:i+1]) for i in range(len(terms))]
    return terms, partial_sums

# Using the geometric sum formula
def geometric_sum_formula(a, r, n):
    return a * (r**(n+1) - 1) / (r - 1)

# Generate terms and partial sums
terms1, sums1 = generate_series_until_limit(a, r1, limit)

# Verify sums with the formula
sum_formula1 = geometric_sum_formula(a, r1, len(terms1) - 1)
print(f"Sum of the series with r={r1} is: {sums1[len(sums1)-1]}")
print(f"Sum of the series with r={r1} using the formula: {sum_formula1}")
```

## Plot and  print the first 20 terms and the partial sums.

```
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(terms1[:20], 'o-', label=f'Terms with r={r1}')
plt.plot(sums1[:20], 'o--', label=f'Partial Sums with r={r1}')

plt.title('First 20 terms and partial sums of geometric series')
plt.xlabel('Term index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Your own example of power sum with $p =$ last digit in the list $+ 1$, i.e., the sum $1^p+2^p+\cdots + n^p$  series up to  $n=10^6$  and verify with the formula.

```
import matplotlib.pyplot as plt
from scipy.special import comb
from mpmath import bernoulli

# Define p
p = 9 + 1

# Faulhaber's formula to compute the sum 1^p + 2^p + ... + n^p
def faulhaber_formula(n, p):
    sum_value = 0
    for k in range(p+1):
        sum_value += comb(p+1, k) * bernoulli(k) * (n**(p+1-k))
    return sum_value / (p + 1)

# Generate series terms and partial sums up to n
def generate_series(p):
    n = 0
    S_n = 0
    partial_sums = []
    terms = []
    while n <= 10**6:
      a_n = n**p
      S_n = S_n + a_n
      partial_sums.append(S_n)
      terms.append(a_n)
      n += 1
    return terms, partial_sums

# Compute terms and partial sums
terms, partial_sums = generate_series(p)



# Verify sum with Faulhaber's formula
sum_faulhaber = faulhaber_formula(n, p)
print(f"Sum up to n={n}: {partial_sums[-1]}")
print(f"Sum up to n={n} using Faulhaber's formula: {sum_faulhaber}")
```
