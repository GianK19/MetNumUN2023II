## SymPyâ€™s Symbolic and Numercical Complex Evaluations




image.png

```
from sympy import symbols, I, N, pi, expand
x1, y1, z1 = symbols("x1 y1 z1 ", real=True)
x = 32 + I*x1
y = 34 + I*y1
z = 39 + I*z1

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))


w = N(1/(pi + I), 20)
print('w =',w)
```
